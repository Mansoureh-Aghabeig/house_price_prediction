# House Price Prediction: Visualization

This notebook demonstrates visualizations for the house price prediction project. We'll load the dataset, perform preprocessing, and use the trained models to make predictions. Visualizations will be used to analyze the relationships in the data and the performance of the models.


### Loading and Preprocessing Data

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys

# Navigate up one level to the project root and then into the 'src' directory
sys.path.append(os.path.abspath('..'))
from src.preprocessing import preprocess_data

# Load the dataset
csv_path = os.path.join('..', 'data', 'USA_Housing.csv')
data = pd.read_csv(csv_path)

# Preprocess the data (splitting into training and test sets)
X_train, X_test, y_train, y_test, scaler = preprocess_data(data)


# Preview the first few rows
data.head()
```

### Data Visualizations

#### Correlation Heatmap

```python
import numpy as np
# Select only numeric columns for correlation matrix
numeric_data = data.select_dtypes(include=[np.number])

# Compute and plot the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=2)
plt.title('Correlation Matrix of Features')
plt.show()
```

### Correlation Heatmap Analysis

The heatmap illustrates the correlation between various features and house prices:

- **Color Scale:**
  - **Dark Red:** Strong positive correlation (-1).
  - **Dark Blue:** Strong negative correlation (+1).
  - **White:** No correlation (0).


- **Key Observations:**
  - **Avg. Area Income and Price:** Strong positive correlation (0.64). Higher income areas have significantly higher house prices
  - **Avg. Area House Age and Price:** Moderate positive correlation (0.45). Older houses tend to have moderately higher prices.
  - **Avg. Area Number of Rooms and Price:** Moderate positive correlation (0.34). More rooms generally lead to higher house prices, but the effect is moderate.
  
  - **Area Population and Price:** Moderate positive correlation (0.41). Higher population density tends to correlate with higher house prices.

This heatmap helps to identify which features are most strongly related to house prices, assisting in feature selection for modeling and further analysis.

#### Scatter Plots

```python
# Scatter plots to explore relationships between features and target variable

plt.figure(figsize=(10, 8))
sns.scatterplot(x='Avg. Area Income', y='Price', data=data)
plt.title('Price vs. Avg. Area Income')
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(10, 8))
sns.scatterplot(x='Avg. Area House Age', y='Price', data=data)
plt.title('Price vs. Avg. Area House Age')
plt.xlabel('Avg. Area House Age')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(10, 8))
sns.scatterplot(x='Avg. Area Number of Rooms', y='Price', data=data)
plt.title('Price vs. Avg. Area Number of Rooms')
plt.xlabel('Avg. Area Number of Rooms')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(10, 8))
sns.scatterplot(x='Avg. Area Number of Bedrooms', y='Price', data=data)
plt.title('Price vs. Avg. Area Number of Bedrooms')
plt.xlabel('Avg. Area Number of Bedrooms')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(10, 8))
sns.scatterplot(x='Area Population', y='Price', data=data)
plt.title('Price vs. Area Population')
plt.xlabel('Area Population')
plt.ylabel('Price')
plt.show()
```

### Scatter Plot Overview
A scatter plot visualizes the relationship between two variables by plotting points on a graph. Each point represents a data sample, with one axis showing the feature (e.g., income or number of rooms) and the other axis showing the house price. The distribution and trend of points help identify the nature of the relationship, such as positive or negative correlations, or whether the data points are scattered without clear trends.

### Feature Interpretations

1. **Avg. Area Income vs. Price**:  
   The scatter plot shows a clear **positive correlation**, where higher average incomes in an area tend to result in higher house prices. The relationship is strong, with most points following an upward trend.

2. **Avg. Area House Age vs. Price**:  
   This scatter plot displays a **moderate positive relationship**. Older houses are somewhat associated with higher prices, but the spread of points indicates that house age alone isn't a strong predictor of price.

3. **Avg. Area Number of Rooms vs. Price**:  
   The scatter plot forms vertical clusters resembling a bar plot, showing that the number of rooms takes on specific values. While more rooms tend to increase prices, there is a wide range of prices for houses with the same number of rooms.

4. **Area Population vs. Price**:  
   This plot suggests a **moderate positive relationship**, where more densely populated areas generally have higher house prices. However, the points are spread out, indicating that other factors also influence the price.


### Distribution of House Prices

```python
# Distribution plot of house prices
plt.figure(figsize=(10, 8))
sns.histplot(data['Price'], bins=30, kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

### Interpretation of House Price Distribution

The distribution of house prices follows a roughly **normal distribution** with the following key observations:

- **Shape**: The histogram presents a bell-shaped curve, suggesting that house prices are normally distributed.

- **Center**: The central tendency (mean) of house prices is around **$1.15 million**.

- **Spread**: Most house prices fall between **$600000  and $1.7 million **, indicating that this is the typical price range for most homes.

- **Skewness**: There is a slight **positive skew** (right tail), meaning that there are fewer houses priced over **$2 million**, but these high-priced homes still exist.

- **KDE (Kernel Density Estimate)**: The smooth line over the histogram (KDE curve) further confirms that the distribution has a single peak and is relatively symmetric, with minor skewness to the right.

This indicates that house prices are generally clustered in a mid-range, with fewer extreme low or high prices.


### Loading Trained Models

```python
import pickle

# Load the trained Linear Regression model
lr_model_path = os.path.join('..', 'models', 'lr_model.pkl')
with open(lr_model_path, 'rb') as file:
    lr_model = pickle.load(file)

# Load the trained Random Forest model (if applicable)
rf_model_path = os.path.join('..', 'models', 'rf_model.pkl')
with open(rf_model_path, 'rb') as file:
    rf_model = pickle.load(file)
```

#### Scatter Plot of Predicted vs Actual Values (Linear Regression)

```python
# Make predictions using the trained linear regression model
predictions = lr_model.predict(X_test)

# Scatter plot of actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, predictions)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Linear Regression: Actual vs Predicted Prices')
plt.show()
```

#### Distribution of Residuals

```python
# Residual plot to visualize prediction errors
residuals = y_test - predictions
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True, color='blue')
plt.title('Distribution of Residuals (Prediction Errors)')
plt.show()
```

## 1. Scatter Plot: Actual vs. Predicted Prices (Linear Regression)

### Interpretation:
- The scatter plot shows a near-linear relationship between the actual house prices (x-axis) and the predicted prices (y-axis).
- The fact that most points align closely with the diagonal line (which represents a perfect prediction) indicates that the linear regression model performs reasonably well.
- There is some dispersion, but it is relatively small, meaning the model predicts house prices with good accuracy, though there are some deviations for higher prices.

## 2. Residual Plot: Distribution of Prediction Errors (Linear Regression)

### Interpretation:
- The histogram represents the distribution of residuals (prediction errors), i.e., the difference between the predicted prices and the actual prices.
- The residuals follow a roughly normal distribution, which is a good sign of a well-fitted linear regression model.
- Most residuals are centered around zero, indicating that the model does not systematically overestimate or underestimate the prices.
- However, there are some larger errors on both ends, suggesting that the model struggles slightly with very high or very low house prices.
